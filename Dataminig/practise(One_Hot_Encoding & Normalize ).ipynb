# **One_Hot_Encoding & Normalize**

```python
import seaborn as sns
import pandas as pd 
import sklearn 
from sklearn.preprocessing import LabelBinarizer
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler

tips_ds = sns.load_dataset('tips')
tips_ds = pd.DataFrame(tips_ds)
one_hot= LabelBinarizer()

sex= tips_ds.loc[:,'sex']
output = one_hot.fit_transform(sex)
tips_ds.drop('sex', axis=1 , inplace=True)
tips_ds = tips_ds.join(pd.DataFrame(output))
tips_ds.rename(columns={0:'sex'},inplace=True)

smoker= tips_ds.loc[:,'smoker']
output = one_hot.fit_transform(smoker)
tips_ds.drop('smoker', axis=1 , inplace=True)
tips_ds = tips_ds.join(pd.DataFrame(output))
tips_ds.rename(columns={0:'smoker'},inplace=True)


time= tips_ds.loc[:,'time']
output = one_hot.fit_transform(time)
tips_ds.drop('time', axis=1 , inplace=True)
tips_ds = tips_ds.join(pd.DataFrame(output))
tips_ds.rename(columns={0:'time'},inplace=True)

# day= tips_ds.loc[:,'day']
# one_hot= LabelBinarizer()
# output = one_hot.fit_transform(day)
# tips_ds.drop('day', axis=1 , inplace=True)
# tips_ds = tips_ds.join(pd.DataFrame(output))
# tips_ds.rename(columns={0:'fri',1:'sat',2:'sun',3:'thur'},inplace=True)   
# AlTERNATIVE >>>
tips_ds=pd.get_dummies(tips_ds)

x = tips_ds.values

#Normalize

#------------------min-max-scaler------------------
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
normalized_df= pd.DataFrame(x_scaled,columns=['total_bill' ,  'tip' , 'size' , 'sex' , 'smoker' , 'time' , 'fri' , 'sat' , 'sun' , 'thur'])
print(normalized_df)

```
