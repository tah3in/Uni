```python
import seaborn as sns
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelBinarizer

tips_ds  = sns.load_dataset('tips')
print(tips_ds.head())

tips_ds = pd.DataFrame(tips_ds)
sex = tips_ds.loc[:,'sex']
one_hot = LabelBinarizer()
output = one_hot.fit_transform(sex)
print(output[:5])
tips_ds.drop('sex', axis=1,inplace=True)
tips_ds = tips_ds.join(pd.DataFrame(output))
# print(tips_ds)
tips_ds.rename(columns={0:'sex'}, inplace=True)

smoker = tips_ds.loc[:,'smoker']
smoker = one_hot.fit_transform(smoker)
tips_ds.drop('smoker', axis=1,inplace=True)
tips_ds = tips_ds.join(pd.DataFrame(smoker))
tips_ds.rename(columns={0:'smoker'}, inplace=True)
print(tips_ds)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
digits = load_digits()
x = digits.data
y = digits.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=1)
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=digits.target_names)
disp.plot()
```

```python
from sklearn.datasets import load_digits, load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix, ConfusionMatrixDisplay
import numpy as np
digits_number = load_iris ()
x = digits_number.data
y = digits_number.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.3, random_state=1)
model = KNeighborsClassifier(n_neighbors=1)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(cm, display_labels=digits_number.target_names)
disp.plot()
```

```python
import matplotlib.pyplot as plt
error = []
for i in range(1,31):
  model = KNeighborsClassifier(n_neighbors=i)
  model.fit(x_train,y_train)
  y_pred = model.predict(x_test)
  error.append(np.mean(y_pred != y_test))

plt.plot(np.linspace(1,30,num=30,endpoint=True),error, marker='*',markerfacecolor='red')
```
